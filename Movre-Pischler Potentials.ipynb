```python
import MarksConstants as mc
```

```python
import sympy as sp
from IPython.display import display
sp.init_printing(use_latex=True)
from importlib import reload
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.style
matplotlib.style.use('dark_background')
import Miscellaneous as misc
```

```python
import IPython.display as displayM
from IPython.display import display
```

```python
Delta, C_3, R = sp.symbols('Delta C_3 R', positive=True, real=True)
```

```python
lifetime = (mc.Rb87_D1_Lifetime+mc.Rb87_D2_Lifetime)/2
transFreq = 2*np.pi*(mc.Rb87_D2LineFrequency+mc.Rb87_D1LineFrequency)/2
DeltaVal = mc.h*(mc.Rb87_D2LineFrequency-mc.Rb87_D1LineFrequency)# / mc.h
#C_3_val = 3*mc.hbar*mc.c**3/(4*lifetime*transFreq**3)# * 1e27 / mc.h
C_3_val = mc.Rb87_C3
print(DeltaVal, C_3_val)
```

```python
V_0_m_g = sp.Matrix([[C_3/R**3 - 2 * Delta / 3, sp.sqrt(2)*Delta/3],[sp.sqrt(2)*Delta/3, -2*C_3/R**3 - Delta / 3]])
V_0_p_g = sp.Matrix([[C_3/R**3 - 2 * Delta / 3, sp.sqrt(2)*Delta/3],[sp.sqrt(2)*Delta/3, 2*C_3/R**3 - Delta / 3]])
V_0_m_u = sp.Matrix([[-C_3/R**3 - 2 * Delta / 3, -sp.sqrt(2)*Delta/3],[-sp.sqrt(2)*Delta/3, -2*C_3/R**3 - Delta / 3]])
V_0_p_u = sp.Matrix([[-C_3/R**3 - 2 * Delta / 3, -sp.sqrt(2)*Delta/3],[-sp.sqrt(2)*Delta/3, 2*C_3/R**3 - Delta / 3]])
V_1_g   = sp.Matrix([[C_3/R**3-Delta/3,-Delta/3,Delta/3],[-Delta/3,-C_3/R**3-Delta/3,Delta/3],[Delta/3,Delta/3,-2*C_3/R**3-Delta/3]])
V_1_u   = sp.Matrix([[-C_3/R**3-Delta/3,-Delta/3,Delta/3],[-Delta/3,C_3/R**3-Delta/3,Delta/3],[Delta/3,Delta/3,2*C_3/R**3-Delta/3]])
display(V_0_m_g, V_0_p_g, V_0_m_u, V_0_p_u, V_1_g, V_1_u)
```

```python
display(displayM.Latex('Expected $0^-_g$ well depth:'))
print(misc.round_sig(DeltaVal/9 * 1e-9 / mc.h,5),'MHz')
display(displayM.Latex('Expected $0^-_g$ well Position:'))
print(misc.round_sig((9*C_3_val/(2*DeltaVal))**(1/3)*1e9),'nm')
```

```python
fig, axs = plt.subplots(2,2,figsize=(20,10))
Rvals = np.linspace( 20*mc.a0, 2000*mc.a0, 3000 )
potentialList = [V_0_m_g, V_0_p_g, V_0_m_u, V_0_p_u, V_1_g, V_1_u]
names=[r'$0^-_g$',r'$0^+_g$',r'$0^-_u$',r'$0^+_u$',r'$1_g$',r'$1_u$', r'$2_g$', r'$2_u$']
linestyles=[':','-','-.',':','-','-.',':','-','-.',':','-','-.']
colors, _ = misc.getColors(len(potentialList)+3, cmStr='viridis')

data = [[] for _ in range(len(potentialList)+2)]

for matrixnum, (name, potentialMatrix, c_) in enumerate(zip(names, potentialList, colors)):
    if matrixnum < 4:
        display(potentialMatrix.eigenvals())
    matrixFunc = sp.lambdify(R, potentialMatrix.subs(C_3,C_3_val).subs(Delta,DeltaVal), 'numpy')
    for rv in Rvals:
        data[matrixnum].append(sorted(np.linalg.eigvals(matrixFunc(rv))))
data[-2] = np.array([C_3_val/rv**3 for rv in Rvals])
data[-1] = np.array([-C_3_val/rv**3 for rv in Rvals])
for ax in axs.flatten():
    ax.axhline(-0.0729113, alpha=0.2,color='r')
    ax.axhline(-0.2298518, alpha=0.2,color='r')
    ax.axhline(-0.30207388, alpha=0.2,color='r')
    ax.axhline(-DeltaVal* 1e-9 / mc.h+0.306246, alpha=0.2,color='r')
    ax.axhline(-DeltaVal* 1e-9 / mc.h-0.510410, alpha=0.2,color='r')
    for name, mdata, c_, linestyle in zip(names, data, colors, linestyles):
        if len(np.array(mdata).shape) == 2 or len(np.array(mdata).shape) == 3:
            for lineNum, lineData in enumerate(misc.transpose(mdata)):
                ax.plot(Rvals*1e9, np.array(lineData) * 1e-9 / mc.h, label=name if lineNum == 0 else None, color=c_, linestyle=linestyle)
        else:
            ax.plot(Rvals*1e9, np.array(mdata) * 1e-9 / mc.h, label=name, color=c_, linestyle=linestyle)
    ax.set_xlabel('Inter-nuclear distance (nm)')
    ax.set_ylabel('Energy (GHz)')
    ax.grid(alpha=0.1)
    ax.axhline(0.1937408, alpha=0.2,color='r', label='F States')

axs[0,0].set_title('Full Range')
axs[0,0].set_xlim(min(Rvals)*1e9,max(Rvals)*1e9/20)
axs[0,0].legend(ncol=5,fontsize=14)
axs[1,0].set_title('Zoom $5P_{3/2}$')
axs[1,0].set_xlim(min(Rvals)*1e9,max(Rvals)*1e9/10)
axs[1,0].set_ylim(-1000, 500)
axs[0,1].set_title('Super-Zoom $5P_{3/2}$')
axs[0,1].set_xlim(min(Rvals)*1e9,max(Rvals)*1e9)
axs[0,1].set_ylim(-0.35, 0.25)
axs[1,1].set_title('Zoom $5P_{1/2}$')
axs[1,1].set_xlim(min(Rvals)*1e9,max(Rvals)*1e9/10)
axs[1,1].set_ylim(-8000, -6000);
```

# With C6 and C8

```python
def RTerms(C3, C6, C8):
    return C3/R**3 - C6/R**6 - C8/R**8
```

```python
Delta, C_3, C6Pi, C6Sig, C8Pi_1, C8Pi_m1, C8Sig_1, C8Sig_m1, R = sp.symbols('Delta C_3  C_6^Pi C_6^Sigma C_8^Pi^1 C_8^Pi^-1 C_8^Sigma^1 C_8^Sigma^-1 R', positive=True, real=True)
```

```python
display(C8Sig_m1, C8Pi_1)
```

```python
V_0_m_g = sp.Matrix([[RTerms(C_3, C6Pi, C8Pi_1) - 2 * Delta / 3,    sp.sqrt(2)*Delta/3],
                     [sp.sqrt(2)*Delta/3,                           RTerms(-2*C_3, C6Sig, C8Sig_1) - Delta / 3]])
V_0_p_g = sp.Matrix([[RTerms(C_3, C6Pi, C8Pi_1) - 2 * Delta / 3,    sp.sqrt(2)*Delta/3],
                     [sp.sqrt(2)*Delta/3,                           RTerms(2*C_3, C6Sig, C8Sig_m1) - Delta / 3]])
V_0_m_u = sp.Matrix([[RTerms(-C_3, C6Pi, C8Pi_m1) - 2 * Delta / 3, -sp.sqrt(2)*Delta/3],
                     [-sp.sqrt(2)*Delta/3,                          RTerms(-2*C_3, C6Sig, C8Sig_1) - Delta / 3]])
V_0_p_u = sp.Matrix([[RTerms(-C_3, C6Pi, C8Pi_m1) - 2 * Delta / 3, -sp.sqrt(2)*Delta/3],
                     [-sp.sqrt(2)*Delta/3,                          RTerms(2*C_3, C6Sig, C8Sig_m1) - Delta / 3]])

V_1_g   = sp.Matrix([[RTerms(C_3, C6Pi, C8Pi_1) - Delta/3,  -Delta/3,  Delta/3],
                     [-Delta/3,  RTerms(-C_3, C6Pi, C8Pi_m1)-Delta/3,  Delta/3],
                     [Delta/3,  Delta/3,  RTerms(-2*C_3, C6Sig, C8Sig_1)-Delta/3]])

V_1_u   = sp.Matrix([[                    RTerms( -C_3, C6Pi,  C8Pi_m1)-Delta/3,  -Delta/3,  Delta/3],
                     [-Delta/3,           RTerms(  C_3, C6Pi,  C8Pi_1)-Delta/3,  Delta/3],
                     [Delta/3,  Delta/3,  RTerms(2*C_3, C6Sig, C8Sig_m1)-Delta/3]])
V_2_g = sp.Matrix([RTerms(C_3, C6Pi, C8Pi_1)])
V_2_u = sp.Matrix([RTerms(-C_3, C6Pi, C8Pi_m1)])
potentialList = [V_0_m_g, V_0_p_g, V_0_m_u, V_0_p_u, V_1_g, V_1_u, V_2_g, V_2_u]
for potential in potentialList:
    display(potential)
```

```python
display(displayM.Latex('Expected $0^-_g$ well depth:'))
print(misc.round_sig(DeltaVal/9 * 1e-9 / mc.h,5),'MHz')
display(displayM.Latex('Expected $0^-_g$ well Position:'))
print(misc.round_sig((9*C_3_val/(2*DeltaVal))**(1/3)*1e9),'nm')
```

```python
Rvals = np.linspace( 1*mc.a0, 2000*mc.a0, 4000 )
names=[r'$0^-_g$',r'$0^+_g$',r'$0^-_u$',r'$0^+_u$',r'$1_g$',r'$1_u$', r'$2_g$', r'$2_u$']
linestyles=[':','-','-.',':','-','-.',':','-','-.',':','-','-.']
colors, _ = misc.getColors(len(potentialList)+3, cmStr='viridis')

data = [[] for _ in range(len(potentialList)+2)]

for matrixnum, (name, potentialMatrix, c_) in enumerate(zip(names, potentialList, colors)):
    evalMatrix = potentialMatrix.subs({Delta:DeltaVal, 
                                       C_3:mc.Rb87_C3,
                                       C6Pi: mc.Rb87_C6Pi, 
                                       C6Sig: mc.Rb87_C6Sigma,
                                       C8Pi_1: mc.Rb87_C8Pi_1,
                                       C8Pi_m1: mc.Rb87_C8Pi_m1,
                                       C8Sig_1: mc.Rb87_C8Sigma_1,
                                       C8Sig_m1: mc.Rb87_C8Sigma_m1})
    matrixFunc = sp.lambdify(R, evalMatrix, 'numpy')
    for rv in Rvals:
        data[matrixnum].append(sorted(np.linalg.eigvals(matrixFunc(rv))))
#data[-2] = np.array([C_3_val/rv**3 for rv in Rvals])
#data[-1] = np.array([-C_3_val/rv**3 for rv in Rvals])
```
